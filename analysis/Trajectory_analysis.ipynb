```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Анализ траектории с и без дисульфидными связями между С3-С40, С4-С32, С16-С26 

RMSD

```python
# with ss-bridges
file_path_with = f'./analysis/withss/rms_with.xvg'
with open(file_path_with, 'r') as file:
    lines = file.readlines()
data_start_line = None
for line_index, line in enumerate(lines):
    if not line.startswith('#') and not line.startswith('@'):
        data_start_line = line_index
        break
rms_with = pd.read_csv(file_path_with, sep=r'\s+', skiprows=data_start_line, header=None)
rms_with.columns = ['Time, ps','RMSD, nm']


# without ss-bridges
file_path_no = f'./no/rms_no.xvg'
with open(file_path_with, 'r') as file:
    lines = file.readlines()
data_start_line = None
for line_index, line in enumerate(lines):
    if not line.startswith('#') and not line.startswith('@'):
        data_start_line = line_index
        break

rms_no = pd.read_csv(file_path_no, sep=r'\s+', skiprows=data_start_line, header=None)
rms_no.columns = ['Time, ps','RMSD, nm']

plt.figure(figsize=(15, 7)) #.set_facecolor('#C8D2D1')
plt.scatter(rms_with['Time, ps'], rms_no['RMSD, nm'], color = '#EE9B01', label='RMSD без ss-мостиков')
plt.axhline(y=rms_no['RMSD, nm'].mean(), color = '#DA6A00', label='Среднее RMSD без ss-мостиков')
plt.scatter(rms_with['Time, ps'], rms_with['RMSD, nm'], color = '#68904D',label='RMSD с ss-мостиков')
plt.axhline(y=rms_with['RMSD, nm'].mean(), color = '#14471E', label='Среднее RMSD с ss-мостиками')
plt.gca().set_facecolor('#C8D2D1')
plt.xlabel('время, пс')
plt.ylabel('RMSD, нм')
plt.legend()
plt.show()
```

RMSF of residues

```python
# with ss-bridges
file_path_with = f'./with/rmsf-per-residue_with.xvg'
with open(file_path_with, 'r') as file:
    lines = file.readlines()
data_start_line = None
for line_index, line in enumerate(lines):
    if not line.startswith('#') and not line.startswith('@'):
        data_start_line = line_index
        break
rms_with = pd.read_csv(file_path_with, sep=r'\s+', skiprows=data_start_line, header=None)
rms_with.columns = ['res_num','RMSD, nm']


# without ss-bridges
file_path_no = f'./no/rmsf-per-residue.xvg'
with open(file_path_with, 'r') as file:
    lines = file.readlines()
data_start_line = None
for line_index, line in enumerate(lines):
    if not line.startswith('#') and not line.startswith('@'):
        data_start_line = line_index
        break

rms_no = pd.read_csv(file_path_no, sep=r'\s+', skiprows=data_start_line, header=None)
rms_no.columns = ['res_num','RMSD, nm']

plt.figure(figsize=(15, 7))
plt.scatter(rms_with['res_num'], rms_no['RMSD, nm'],color = '#EE9B01', label='среднее RMSF аминокислотных остатков без ss-мостиков')
plt.scatter(rms_with['res_num'], rms_with['RMSD, nm'],color = '#68904D', label='среднее RMSF аминокислотных остатков с ss-мостиков')
plt.axvline(x=3, color = 'red', label='CYS 3')
plt.axvline(x=4, color = 'blue', label='CYS 4')
plt.axvline(x=16, color = 'green', label='CYS 16')
plt.axvline(x=26, color = 'magenta', label='CYS 26')
plt.axvline(x=32, color = 'yellow', label='CYS 33')
plt.axvline(x=40, color = 'grey', label='CYS 40')
plt.gca().set_facecolor('#C8D2D1')
plt.xlabel('№ аминокислотного остатка')
plt.ylabel('RMSF, нм')
plt.legend()
plt.show()
```

Вдородные связи внутри белка.

```python
# with ss-bridges
file_path_with = f'./with/hboond_with_pr.xvg'
with open(file_path_with, 'r') as file:
    lines = file.readlines()
data_start_line = None
for line_index, line in enumerate(lines):
    if not line.startswith('#') and not line.startswith('@'):
        data_start_line = line_index
        break
rms_with = pd.read_csv(file_path_with, sep=r'\s+', skiprows=data_start_line, header=None)
rms_with.drop(columns=rms_with.columns[2], inplace = True)
rms_with.columns = ['Time, ps','hbonds']


# without ss-bridges
file_path_no = f'./no/hboond_no_pr.xvg'
with open(file_path_with, 'r') as file:
    lines = file.readlines()
data_start_line = None
for line_index, line in enumerate(lines):
    if not line.startswith('#') and not line.startswith('@'):
        data_start_line = line_index
        break

rms_no = pd.read_csv(file_path_no, sep=r'\s+', skiprows=data_start_line, header=None)
rms_no.drop(columns=rms_no.columns[2], inplace = True)
rms_no.columns =  ['Time, ps','hbonds']

plt.figure(figsize=(15, 7))
plt.scatter(rms_with['Time, ps'], rms_no['hbonds'], label='Количество водородных связей внутри белка без ss-мостиков')
plt.scatter(rms_with['Time, ps'], rms_with['hbonds'], label='Количество водородных связей внутри белка с ss-мостиков')
plt.axhline(y=rms_with['hbonds'].mean(), color = 'red', label='Среднее количество водородных связей с ss-мостиков')
plt.axhline(y=rms_no['hbonds'].mean(), color = 'green', label='Среднее количество водородных связей без ss-мостиков')
plt.xlabel('RMSD, nm')
plt.ylabel('Time, ps')
plt.legend()
plt.show()
```
